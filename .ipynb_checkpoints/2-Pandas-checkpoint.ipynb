# **Series**

```python
#https://jalammar.github.io/gentle-visual-intro-to-data-analysis-python-pandas/
import numpy as np
import pandas as pd
```

```python
grades = {"Atil": 50, "James": 60, "Lars": 30}
```

```python
pd.Series(grades)
```

```python
names = ["Atil", "James", "Lars"]
grades = [50,60,30]
```

```python
pd.Series(names)
```

```python
pd.Series(grades)
```

```python
pd.Series(names, grades)
```

```python
pd.Series(data=names, index=grades)
```

```python
numpy_array = np.array([50,40,30,20])
pd.Series(numpy_array)
```

```python
contest_result1 = pd.Series(data=[10,5,100], index=["Atil","James","Lars"])
contest_result2 = pd.Series(data=[20,50,10], index=["Atil","James","Lars"])
```

```python
#multi  index
```

```python
first_index = ["simpson","simpson","simpson","south park","south park","south park"]
inner_index = ["homer","bart","marge","cartman","kenny","kyle"]
```

```python
zipped = list(zip(first_index, inner_index))
zipped
```

```python
zipped = pd.MultiIndex.from_tuples(zipped)
zipped
```

```python
sample_values = np.ones((6,2))
```

```python
zip_pd = pd.DataFrame(sample_values, index=zipped, columns=["age","salary"])
zip_pd
```

```python
zip_pd.loc["simpson"]
```

```python
zip_pd.loc["simpson"].loc["homer"]
```
