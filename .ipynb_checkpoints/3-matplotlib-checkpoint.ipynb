```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
age_list=[10,20,30,30,30,40,50,60,70,75]
weight_list=[20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list, "red")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("weight-age")
plt.show()
#plt.savefig("test.png")
```

```python
#numpy
```

```python
np_age_list = np.array(age_list)
np_wight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_wight_list, "purple")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("weight-age")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20) #1den 10a kadar eşit aralıklarla 20 sayı yazar
numpy_arr1 
```

```python
numpy_arr2 = numpy_arr1**3
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "g*-") # g, g-, g*, g*-, g--
plt.show()
```

```python
plt.plot(numpy_arr2, numpy_arr1, "g-") # g, g-, g*, g*-, g--
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr2, numpy_arr1, "g-")
plt.subplot(1,2,2)
plt.plot(numpy_arr1, numpy_arr2, "b--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.4,0.4])
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("x axis")
figure_axes.set_ylabel("y axis")
figure_axes.set_title("graph")
plt.show()
```

```python
new_fig = plt.figure(dpi=200)

new_axes = new_fig.add_axes([0.1, 0.1, 0.9, 0.9])

new_axes.plot(numpy_arr1, numpy_arr1**2, label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label = "numpy array ** 3")
new_axes.legend(loc = 2)
plt.show()
```

```python
data1 = np.linspace(0,10,20)
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color="#1E0704", alpha=0.5)
my_axes.plot(data2, data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1, data1+2, color="blue", linewidth=1.0)
new_axes.plot(data1, data1+3, color="yellow", linewidth=5.0)

new_axes.plot(data1, data1+4, color="black", linestyle="-.")
new_axes.plot(data1, data1+5, color="green", linestyle=":")
new_axes.plot(data1, data1+6, color="purple", linestyle="--")

new_axes.plot(data1, data1+7, color="black", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1+8, color="black", linestyle="--", marker="+", markersize=9)

plt.show()
```

```python
#scatter
```
