```python
# z-score
```

```python
data = np.array([10,12,13,15,18,25,100,105])
```

```python
#outlier
```

```python
mean = np.mean(data)
mean
```

```python
data.sum() / len(data)
```

```python
std = np.std(data)
std
```

```python
z_score = (data-mean) / std
z_score
```

```python
z_score[z_score > 1]
```

```python
data[z_score > 1]
```
